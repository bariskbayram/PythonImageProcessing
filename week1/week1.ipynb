```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = plt.imread('images.jpeg')
plt.imshow(image1)
plt.show()
```

```python
image2 = image1[:,:,0]
plt.imshow(image2, cmap='gray')
plt.show()
```

```python
def photoBlackWhite(image): #Aynı zamanda fotoğraf 2Dimension hale gelir
    m,n,p = image.shape
    newImage = np.zeros((m,n), dtype=int)
    for i in range(m):
        for j in range(n):
            s = int((image[i,j,0]+image[i,j,1]+image[i,j,2])/3)
            newImage[i,j] = s
    return newImage
```

```python
image3 = photoBlackWhite(image1)
plt.imshow(image3, cmap='gray')
plt.show()
```

```python
def photoBlackWhiteFor2Dimension(image): #2Dimension olarak gelen fotoğraf sadece siyah beyaz olarak çıkıyor
    m,n = image.shape
    newImage = np.zeros((m,n), dtype=int)
    for i in range(m):
        for j in range(n):
            s = image[i,j]
            newImage[i,j] = s
    return newImage
```

```python
image4 = photoBlackWhiteFor2Dimension(image2)
plt.imshow(image4, cmap='gray')
plt.show()
```

```python
def transpozeSpatialFor3Dimension(image): #3Dimension olarak gelir, çıkan fotoğraf 2 boyutlu hale gelir
    m,n,p = image.shape
    newImage = np.zeros((n,m), dtype=int)
    for i in range(m):
        for j in range(n):
            s = int((image[i,j,0]+image[i,j,1]+image[i,j,2])/3)
            newImage[j,i] = s
    return newImage
```

```python
image5 = transpozeSpatialFor3Dimension(image1)
plt.imshow(image5, cmap='gray')
plt.show()
```

```python
def transpozeSpatialForAnyDimension(image): #Dimension sabit kalarak transpoze işlemi
    flag = True
    if(len(image.shape) == 2):
        m,n = image.shape
        newImage = np.zeros((n,m), dtype=int)
    else:
        m,n,p = image.shape
        newImage = np.zeros((n,m,p), dtype=int)
        flag = False
    for i in range(m):
        for j in range(n):
            if(flag == False):
                newImage[j,i,0] = image[i,j,0]
                newImage[j,i,1] = image[i,j,1]
                newImage[j,i,2] = image[i,j,2]
            else:
                newImage[j,i] = image[i,j]
    return newImage
```

```python
image6 = transpozeSpatialForAnyDimension(image1) #3Dimension fotoğraf geldi transpoze olup döndü
plt.imshow(image6)
plt.show()
```

```python
image7 = transpozeSpatialForAnyDimension(image4) #2Dimension fotoğraf geldi transpoze olup döndü
plt.imshow(image7, cmap='gray')
plt.show()
```

```python
def transposeMirror3DimensionTo2Dimension(image):
    m,n = image.shape
    newImage = np.zeros((m,n), dtype=int)
    for i in range(m):
        for j in range(n):
            newImage[i, n-1-j] = image[i,j]
    return newImage
```

```python
image11 = transposeMirror3DimensionTo2Dimension(image2)
plt.imshow(image11, cmap='gray')
plt.show()
```
