```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def calculateMeanAndVar(myList=[2,3,4,5,2,1,8,6,5,4]):
    t=0
    s=0
    for i in myList:
        s=s+1
        t=t+i
    mean=t/s
    
    t=0
    s=0
    for i in myList:
        s=s+1
        t=t+(i-mean)*(i-mean)
    var=t/(s-1)
    var=math.sqrt(var)
    
    return mean, var
```

```python
calculateMeanAndVar()
```

```python
def creatingHistogram(myList=[2,3,4,5,2,1,8,6,5,4]):
    histogram = {}
    for i in myList:
        if i in histogram.keys():
            histogram[i] = histogram[i]+1
        else:
            histogram[i] = 1
    print(histogram)
```

```python
creatingHistogram()
```

```python
def creatingHistogramOfPhoto(image):
    m,n,p = image.shape
    photoHistogram = {}
    for i in range(m):
        for j in range(n):
            for k in range(p):
                if image[i,j,k] in photoHistogram.keys():
                    photoHistogram[image[i,j,k]] += 1
                else:
                    photoHistogram[image[i,j,k]] = 1
    return photoHistogram
```

```python
image1 = plt.imread('images.jpeg')
plt.imshow(image1)
plt.show()
```

```python
creatingHistogramOfPhoto(image1)
```

```python
def covertToRealHistogram(histogram):
    x=[]
    y=[]
    for key in histogram.keys():
        x.append(key)
        y.append(histogram[key])
    return x,y,histogram
```

```python
x, y, histogram = covertToRealHistogram(creatingHistogramOfPhoto(image1))
plt.bar(x,y)
plt.show()
```
