```python
image1 = plt.imread('images.jpeg')
plt.imshow(image1)
plt.show()
```

```python
def intensityForEachRGBValue(image, r=0, g=0, b=0):
    m,n,p = image.shape
    newImage = np.zeros((m,n,p), dtype=int)
    for i in range(m):
        for j in range(n):
            newImage[i,j,0] = image[i,j,0] + r
            newImage[i,j,1] = image[i,j,1] + g
            newImage[i,j,2] = image[i,j,2] + b
    return newImage
```

```python
image2 = intensityForEachRGBValue(image1, 80, 10, 20)
plt.imshow(image2)
plt.show()
```

```python
def scaleDown(image): #znrl yarya dr ve 2Dimension yap
    m,n,p = image.shape
    newM = int(m/2)
    newN = int(n/2)
    newImage = np.zeros((newM,newN), dtype=int)
    for i in range(newM):
        for j in range(newN):
            s = int((image[i*2,j*2,0]+image[i*2,j*2,1]+image[i*2,j*2,2])/3)
            newImage[i,j] = s
    return newImage
```

```python
image3 = scaleDown(image1)
plt.imshow(image3, cmap='gray')
plt.show()
```

```python
def scaleDownForRGB(image): #Hem znrl yarya dryor hem de 3Dimension kalyor
    m,n,p = image.shape
    newM = int(m/2)
    newN = int(n/2)
    newImage = np.zeros((newM, newN,3), dtype=int)
    for i in range(newM):
        for j in range(newN):
            newImage[i,j,0] = image[i*2,j*2,0]
            newImage[i,j,1] = image[i*2,j*2,1]
            newImage[i,j,2] = image[i*2,j*2,2]
    return newImage
```

```python
image4 = scaleDownForRGB(image1)
plt.imshow(image4)
plt.show()
image5 = scaleDownForRGB(image4)
plt.imshow(image5)
plt.show()
```

```python
def scaleDown3(image): #Fotorafn orjinal boyutunun 1/4 ksmna fotorafn kltlm hali konur
    m,n,p = image.shape #Bunu her 4 ke iin yapar
    newM = int(m/2)
    newN = int(n/2)
    newImage = np.zeros((m,n,3), dtype=int)
    for i in range(newM):
        for j in range(newN):
            newImage[i,j,0] = image[i*2,j*2,0]
            newImage[i,j,1] = image[i*2,j*2,1]
            newImage[i,j,2] = image[i*2,j*2,2]
    for i in range(newM):
        for j in range(newN):
            newImage[i+newM,j,0] = image[i*2,j*2,0]
            newImage[i+newM,j,1] = image[i*2,j*2,1]
            newImage[i+newM,j,2] = image[i*2,j*2,2]
    for i in range(newM):
        for j in range(newN):
            newImage[i+newM,j+newN,0] = image[i*2,j*2,0]
            newImage[i+newM,j+newN,1] = image[i*2,j*2,1]
            newImage[i+newM,j+newN,2] = image[i*2,j*2,2]
    for i in range(newM):
        for j in range(newN):
            newImage[i,j+newN,0] = image[i*2,j*2,0]
            newImage[i,j+newN,1] = image[i*2,j*2,1]
            newImage[i,j+newN,2] = image[i*2,j*2,2]
            
    return newImage
```

```python
image7 = scaleDown3(image1)
plt.imshow(image7)
plt.show()
```

```python
def inverseIntensity(image): #Fotorafn younluklarnn tersini alr
    m, n, p = image.shape
    newImage = np.zeros((m, n, 3), dtype=int)

    for m in range(image.shape[0]):
        for n in range(image.shape[1]):
            newImage[m, n, 0] = 255 - image[m, n, 0]
            newImage[m, n, 1] = 255 - image[m, n, 1]
            newImage[m, n, 2] = 255 - image[m, n, 2]

    return newImage
```

```python
image8 = inverseIntensity(image1)
plt.imshow(image8)
plt.show()
```
