```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image1 = plt.imread('images.jpeg')
plt.imshow(image1)
plt.show()
m,n,p = image1.shape
```

```python
image2 = np.zeros((m, n), dtype=np.uint8)
image2 = image1[:,:,0]
plt.imshow(image2, cmap='gray')
plt.show()
```

```python
m,n = image2.shape
image3 = image1[:,:,0] # image1=>gerçek fotoğraf, image2=>filter uygulanacak, image3=>filter uygulandıktan sonra
image3 = np.zeros((m,n), dtype=np.uint8)
for i in range(1, m-1):
    for j in range(1, n-1):
        s = image2[i-1,j-1]/9 + image2[i-1,j]/9 + image2[i-1,j+1]/9 + image2[i,j-1]/9 + image2[i,j]/9+ \
        image2[i,j+1]/9 + image2[i+1,j-1]/9 + image2[i+1,j]/9 + image2[i+1,j+1]/9 
        s = int(s)
        image3[i,j] = s
        
plt.imshow(image3, cmap='gray')
plt.show()
```
