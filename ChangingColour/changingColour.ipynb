```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = plt.imread("motor.jpg")
plt.imshow(image)
plt.show()
```

```python
def changingColour(image, red, green, blue): #kırmızı için red = 61,255 ve green 0,164 ve blue 0,164
    m,n,p = image.shape
    output = np.zeros((m,n,p), dtype=int)
    redMin, redMax = red[0], red[1]
    greenMin, greenMax = green[0], green[1]
    blueMin, blueMax = blue[0], blue[1]
    print(image[250,300,:])
    for i in range(m):
        for j in range(n):
            if(image[i,j,0]>=redMin and image[i,j,0]<=redMax):
                if(image[i,j,1]>=greenMin and image[i,j,1]<=greenMax):
                    if(image[i,j,2]>=blueMin and image[i,j,2]<=blueMax):
                        if((image[i,j,0] - image[i,j,1]) > 50):
                                output[i,j,0] = 31
                                output[i,j,1] = 190
                                output[i,j,2] = 220
                        else:
                            output[i,j,:] = image[i,j,:]
                    else:
                        output[i,j,:] = image[i,j,:]
                else:
                    output[i,j,:] = image[i,j,:]
            else:
                output[i,j,:] = image[i,j,:]
    return output
```

```python
newImage = changingColour(image, red=[70,255], green=[0,130], blue=[0,130])
plt.imshow(newImage)
plt.show()
```
