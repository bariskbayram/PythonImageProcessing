```python
def myDistance(a, b):
    myPoint1 = a
    myPoint2 = b
    a = (myPoint1[0]-myPoint2[0])**2
    b = (myPoint1[1]-myPoint2[1])**2
    return math.sqrt(a+b)
```

```python
def findCenter(ListOfPoints):
    a=0
    b=0 
    length = len(ListOfPoints)
    for point in ListOfPoints:
        a = a + point[0]
        b = b + point[1]
    return [a/length,b/length]
```

```python
im1=[[1,0],[0,0]]
im2=[[0,1],[0,0]]
im3=[[0,0],[1,0]]
im4=[[0,0],[0,1]]
em1=[[0,1],[1,1]]
em2=[[1,0],[1,1]]
em3=[[1,1],[0,1]]
em4=[[1,1],[1,0]]
im1 = [im1, im2, im3, im4]
em1 = [em1, em2, em3, em4]

img1 = plt.imread("images.jpeg")
plt.imshow(img1, plt.cm.binary)
plt.show()

def count_internal_mask(image):
    counter_internal = 0
    for mask in im1:
        counter_internal = counter_internal + count_mask(image, mask)
    return counter_internal
def count_external_mask(image):
    counter_external = 0
    for mask in em1:
        counter_external = counter_external + count_mask(image, mask)
    return counter_external
def count_mask(image, mask):
    counter = 0
    m = image.shape[0]
    n = image.shape[1]
    for i in range(m-1):
        for j in range(n-1):
            a = b = c = d = False
            if(image[i,j][0] == mask[0][0]):
                a = True
            if (image[i,j+1][0] == mask[0][1]):
                b = True
            if (image[i+1,j][0] == mask[1][0]):
                c = True
            if (image[i+1,j+1][0] == mask[1][1]):
                d = True
            if (a and b and c and d):
                counter = counter + 1
    return counter
```

```python
c1 = count_internal_mask(img1)
c2 = count_external_mask(img1)
c1, c2, count_mask(img1, im2)
```

```python
#(Yellow, Filter)

m,n,p = img1.shape
image1 = np.zeros((m,n,3), dtype=int)
image1[:,:,:] = img1[:,:,:]

image1[:,:,2]=0.0

plt.imshow(image1)

plt.show()
```

```python
#(Round, Filter)

im_size=img1.shape
center=[im_size[0]/2,im_size[1]/2]

for i in range(im_size[0]):
    for j in range(im_size[1]):
        if (myDistance([i,j],center)>100):
            image1[i,j,:]=0
plt.imshow(image1)
plt.show()
```
