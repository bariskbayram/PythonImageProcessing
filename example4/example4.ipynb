```python
import numpy as np
import matplotlib.pyplot as plt
import random as random
```

```python
image = plt.imread('images.jpeg')
plt.imshow(image)
plt.show()
```

```python
def get_default_mask_for_mean():
    return np.array([1,1,1,1,1,1,1,1,1]).reshape(3,3)/9
```

```python
def apply_mask(part_of_image):
    mask = get_default_mask_for_mean()
    return np.average(part_of_image) #aslnda mask uygulanmyor nk bu mask sadece ortalama alyor
```

```python
def covert_rgb_to_gray_level(image):
    m,n,p = image.shape
    output = np.zeros((m,n), dtype=int)
    for i in range(m):
        for j in range(n):
            output[i,j] = image[i,j,0]/3 + image[i,j,1]/3 + image[i,j,2]/3
    return output
```

```python
image1 = covert_rgb_to_gray_level(image)
plt.imshow(image1, cmap='gray')
plt.show()
```

```python
m,n  = image1.shape
output = np.zeros((m,n), dtype=int)
for i in range(1,m):
    for j in range(1,n):
        poi = image1[i-1:i+2, j-1:j+2]
        output[i,j] = apply_mask(poi)
plt.imshow(output, cmap='gray')
plt.show()
```

```python
def get_median(poi):
    s_1=poi.reshape(1,9)
    s_1.sort()
    return s_1[0,4]
```

```python
output2 = np.zeros((m,n), dtype=int)
for i in range(1,m-1):
    for j in range(1,n-1):
        poi = image1[i-1:i+2,j-1:j+2]
        output2[i,j] = get_median(poi)
plt.imshow(output2, cmap='gray')
plt.show()
```
